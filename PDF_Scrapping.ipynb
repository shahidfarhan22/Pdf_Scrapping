The url input for downloading pdfs and setting up output folder path

```python
# Save contenets from url into folder_location
url="https://www.premierleague.com/about/publications"
folder_location='/content/drive/MyDrive/Colab Notebooks/Premier_league'
if not os.path.exists(folder_location):
  os.mkdir(folder_location)
```

```python
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')

#loop through all PDF links in the page
for link in soup.select("a[href$='.pdf']"):
  #local file name is same as the PDF file name in the URL (ignore rest of the path)
  filename= os.path.join(folder_location,link['href'].split('/')[-1])

  with open(filename,'wb') as f:
    f.write(requests.get(urljoin(url,link['href'])).content)
```

Reading a table from a PDF document and storing it in a csv file

```python
combined_pdf=folder_location+"/PL_Handbook_2023-24_DIGITAL_21.07.23.pdf"
tabula.read_pdf(combined_pdf,pages='360')
```

```python
from tabula import convert_into
convert_into(combined_pdf,folder_location+"/table_output.csv",output_format='csv',pages=360)
pd.read_csv(folder_location+"/table_output.csv")
```
